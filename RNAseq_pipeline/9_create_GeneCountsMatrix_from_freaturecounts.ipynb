```python
df = pd.read_csv("9_create_GeneCountsMatrix/gene_matrix_count.csv", header=0,
                 sep=',', index_col='Geneid')
df
```

```python
# Define a function to extract the desired substring
def extract_sample_name(column_name):
    # Find the last occurrence of '/' and '.'
    last_slash = column_name.rfind('/')
    last_dot = column_name.rfind('.')

    # Extract the substring between the last '/' and the last '.'
    sample_name = column_name[last_slash + 1:last_dot]

    return sample_name

# Apply the function to each column and rename the columns
df.columns = df.columns.map(extract_sample_name)

# remove 24 from all column containing an R 
df.columns = df.columns.str.replace('24R', 'R')

# Display the resulting DataFrame with renamed columns
print(df)

# export the dataframe to a csv file
df.to_csv("9_create_GeneCountsMatrix/gene_matrix_count.csv")
```

# Create table to sample information 

```python
import pandas as pd

# Extract conditions from column names
conditions = []
for column in df.columns:
    if 'R' in column:
        conditions.append('reoxygenation')
    elif 'A' in column:
        conditions.append('anoxia')
    elif 'N' in column:
        conditions.append('normoxia')
    else:
        conditions.append('unknown')  # or handle other cases

# Create a new DataFrame
sample_info = pd.DataFrame({'Geneid': df.columns, 'condition': conditions[:len(df.columns)]})

# Display the resulting DataFrame
print(sample_info)

# export to csv
sample_info.to_csv('9_create_GeneCountsMatrix/sample_info.csv', index=False)
```
