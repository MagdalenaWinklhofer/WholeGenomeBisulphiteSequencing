```python
# Import packages 
import pandas as pd
import numpy as np
from statsmodels.stats.multitest import multipletests
```

# 1) Import GO enrichement output (Rstudio)

```python
# Import output from GO enrichment analysis in Rstudio 
enrichment = pd.read_csv('/home/maggy/WholeGenomeBisulphiteSequencing/GO_analysis/2_filter_GO_erichement_analysis/NvsA_GOterms.txt', sep='\t')

# Import subset (for caluclating corrections)
subset = pd.read_csv('/home/maggy/WholeGenomeBisulphiteSequencing/GO_analysis/1_prepare_data_for_GO_analysis/goseq_NvsA.csv', sep=',', index_col=0)

# import the gene lenth df to get the geneids for the background 
background = pd.read_csv('/home/maggy/WholeGenomeBisulphiteSequencing/GO_analysis/1_prepare_data_for_GO_analysis/gene_length.csv', sep=',')
```

```python

# calculate expected number in each category 
enrichment['expectedInCat'] = (enrichment['numDEInCat']/50000)* 35
# 50000 number of GO terms 


# calculate fold enrichment
enrichment['fold_enrichment'] = enrichment['numDEInCat']/enrichment['expectedInCat']

# calculate negative fold change from the FDR corrected values
enrichment['-log10(FDR)'] = -np.log10(enrichment['over_represented_pvalue'])

```

```python
import statsmodels.stats.multitest as smm
enrichment['corrected_p_values'] = smm.multipletests(enrichment['over_represented_pvalue'], method='bonferroni')[1]


import statsmodels.stats.multitest as smm

enrichment['q_values'] = smm.multipletests(enrichment['over_represented_pvalue'], method='fdr_bh')[1]

```

```python
import statsmodels.stats.multitest as smm

enrichment['corrected_p_values'] = smm.multipletests(enrichment['over_represented_pvalue'], alpha=0.05, method='hs')[1]
```
